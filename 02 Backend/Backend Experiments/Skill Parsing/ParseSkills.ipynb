```python
import json
```

```python
with open("Skills.txt", "rt") as file:
    file_contents = file.readlines()
```

```python
for line in file_contents:
    if line == '\n':
        continue

    else:
        print(line)
```

```python
skills = {}
```

```python
for line in file_contents:
    if line == '\n':
        continue

    else:
        split_line = line.split("::")

        # Make sure all skills have a primary and a secondary category (Along with the Skill)
        assert len(split_line) == 3, split_line

        primary_category, secondary_category, skill = split_line[0], split_line[1], split_line[2]

        # If the primary category does not already exist in the dictionary, create it now.
        if not (primary_category in skills):
            skills[primary_category] = {}

        # If the secondary category does not already exist in the dictionary, create it now.
        if not (secondary_category in skills[primary_category]):
            skills[primary_category][secondary_category] = []

        skills[primary_category][secondary_category].append(skill.replace("\n", ""))
        
```

```python
skills
```

```python
with open("Skills.json", "wt") as file:
    json.dump(skills, file)
```
